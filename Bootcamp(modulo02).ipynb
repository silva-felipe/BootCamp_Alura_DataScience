## Desafio 02 - Padronizar e transformar
- Padronizar os nomes das colunas
- Transformar os dados nos tipos corretos
- Transformar a UF em index

limpando os dados

```python
pop_estados = pd.read_html(url)[0]
```

```python
pop_estados.head()
```

retirando as colunas indesejadas

```python
pop_estados = pop_estados.drop(['Posição','País comparável (habitantes)','% da pop. total'], axis=1)
```

```python
pop_estados.head()
```

verificando quais sao as colunas existentes e renomeando-as

```python
pop_estados.columns
```

```python
pop_estados = pop_estados.rename(columns={pop_estados.columns[0]:'uf', pop_estados.columns[1]:'populacao'})
```

```python
pop_estados.head()
```

setando um novo index

```python
pop_estados = pop_estados.set_index(pop_estados.columns[0])
```

```python
pop_estados.head()
```

retirando os espaços dos valores na coluna 'populacao' utilizando a função replace de python junto com a .str do pandas, que permite utilizar algumas funções de string do python

```python
pop_estados['populacao'] = pop_estados['populacao'].str.replace(' ','')
```

```python
pop_estados.head()
```

verificando os tipos dos dados, observando que todos são do tipo 'object'

```python
pop_estados.info()
```

transformando os valores de 'object' para 'int64' utilizando a função 'to_numeric'
fonte: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_numeric.html#pandas.to_numeric

```python
pop_estados['populacao'] = pd.to_numeric(pop_estados['populacao'])
```

```python
pop_estados.head()
```

verificando os tipos dos dados, observado que após a utilização da função .to_numeric, agora os dados são do tipo int64

```python
pop_estados.info()
```

## Desafio 04 -  Desenvolver uma função para a limpeza e tratamento dos dados, retornando duas tabelas "população" e "unida"

```python
import pandas as pd
pd.options.display.float_format = "{:.2f}".format

def populacao_estados():
  url = input('digite url: ')
  url = 'http:' + url[6:]
  pop_estados = pd.read_html(url)[0]
  pop_estados = pop_estados.drop(['Posição','País comparável (habitantes)','% da pop. total'], axis=1)
  pop_estados = pop_estados.rename(columns={pop_estados.columns[0]:'uf', pop_estados.columns[1]:'populacao'})
  pop_estados = pop_estados.set_index(pop_estados.columns[0])
  pop_estados['populacao'] = pop_estados['populacao'].str.replace(' ','')
  pop_estados['populacao'] = pd.to_numeric(pop_estados['populacao'])
  

  url_processamento = input('digite url: ')
  df_processamento = pd.read_csv(url_processamento, encoding='ISO-8859-1',sep=';',skiprows=3,skipfooter=12,thousands='.',decimal=',')
  df_processamento['uf'] = df_processamento['Unidade da Federação'].str[3:]
  df_processamento.drop(['Unidade da Federação','Total'], axis=1, inplace=True)
  df_processamento.set_index('uf', inplace=True)
  gasto_do_mais_recente = df_processamento[df_processamento.columns[-1]]

  pop_gastos = pop_estados.join(gasto_do_mais_recente)
  
  return display(pop_estados), display(pop_gastos)


populacao_estados()
```
