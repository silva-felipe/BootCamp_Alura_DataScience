# Video 03 - Aula 1

**recriando a coluna total**

```python
usaveis['Total'] = usaveis.sum(axis=1)
```

**Deasafio: acrescentar uma coluna com região**

```python
index = usaveis.index.tolist()
index
```

```python
regiao = []
for i in index:
  if int(i[:2]) < 20:
    regiao.append('Norte')
  elif 20 < int(i[:2]) < 30:
    regiao.append('Nordeste')
  elif 30 < int(i[:2]) < 40:
    regiao.append('Sudeste')
  elif 40 < int(i[:2]) < 50:
    regiao.append('Sul')
  else:
    regiao.append('Centro-Oeste')

regiao
```

```python
usaveis['Região'] = regiao
```

```python
usaveis.head()
```

```python
usaveis[['Região']]
```

**Selecionando os estados de São Paulo e Bahia para fazer uma comparação de gasto per capita em cada estado**

```python
usaveis.head()
```

Selecionando as colunas que irao ser plotadas

Calculando o gasto per capita dos estados

```python
gosto_per_capita_sp = sp_ba.loc['35 São Paulo'] / pop_est_sp_2007
```

```python
gasto_per_capita_ba = sp_ba.loc['29 Bahia'] / pop_est_ba_2007
```

reatribuindo os valores pelo gasto per capita no dataframe

Plotando o valor de gasto per capita dos estados de São Paulo e Bahia no período de Fev/2008 a Jan 2021, utilizando como base a população do ultimo censo realizado (2007) conseguido no site do IBGE

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
```

```python
ax9 = sp_ba.T.plot(figsize=(18,9))
ax9.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.0f}"))
loc = plticker.MultipleLocator(base=1.0)
ax9.yaxis.set_major_locator(loc)
plt.xticks(rotation=85)
plt.ylim(0,10)
plt.title('Gasto per capita dos estados de São Paulo e Bahia\n de fevereiro/2008 a janeiro/2021\n'.upper(), fontsize=18, weight='bold')
plt.xlabel('\nAno/Mês', fontsize=15)
plt.grid(color='r', linestyle=':', linewidth=0.5, alpha=0.7)
plt.show()
```
