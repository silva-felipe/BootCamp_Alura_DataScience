# Contextualização

**COBERTURA VACINAL**

A cobertura vacinal é o percentual de crianças imunizadas por vacinas específicas, dentro de um espaço geográfico delimitado em um determinado ano.

• Estima o nível de proteção da população infantil contra doenças selecionadas, evitáveis por imunização, mediante o cumprimento do esquema básico de vacinação.
<b>fonte:</b> *http://www.ripsa.org.br/fichasIDB/pdf/ficha_F.13.pdf*

**VACINAS ANALISADAS**

- <b>BCG</b> (<b>fonte:</b> *https://familia.sbim.org.br/vacinas/vacinas-disponiveis/vacina-bcg*)
 - Previne contra Tuberculose – principalmente as formas graves, como meningite tuberculosa e tuberculose miliar (espalhada pelo corpo).
 - É composta pelo bacilo de Calmette-Guérin – origem do nome BCG – obtido pela atenuação (enfraquecimento) de uma das bactérias que causam a tuberculose. Completam sua composição o glutamato de sódio e a solução fisiológica (soro a 0,9%).

  - <b>Indicação:</b> 

   - A vacina é indicada de rotina a partir do nascimento até antes de a criança completar 5 anos de idade.

- <b>Rotavírus humano</b> (<b>fonte:</b> *https://familia.sbim.org.br/vacinas/vacinas-disponiveis/vacina-rotavirus*)

 - Previne contra doença diarreica causada por rotavírus.
 - Vacina oral monovalente (VRH1) contém um tipo de rotavírus vivo “enfraquecido”, além de sacarose, adipatodissódico, meio Eagle modificado Dulbecco (DMEM) e água estéril.

 - Vacina oral atenuada pentavalente (VRH5) é composta por cinco tipos de rotavírus vivos “enfraquecidos”, sacarose, citrato de sódio, fosfato de sódio monobásico monoidratado, hidróxido de sódio, polissorbato 80, meios de cultura e traços de soro fetal bovino.

 - <b>Indicação:</b>

   - Bebês de 6 semanas a 8 meses e 0 dia. A primeira dose deve ser obrigatoriamente aplicada até a idade de 3 meses e 15 dias, e a última dose até os 7 meses e 29 dias.


 - <b>Meningocócica B</b> (<b>fonte:</b> *https://familia.sbim.org.br/vacinas/vacinas-disponiveis/vacina-meningococica-b*)

  - Previne contra meningites e infecções generalizadas (doenças meningocócicas) causadas pela bactéria meningococo do tipo B.

  - Trata-se de vacina inativada, portanto, não causa infecção.

  - É composta por quatro componentes (três proteínas subcapsulares e vesículas da membrana externa do meningococo B), além de hidróxido de alumínio, cloreto de sódio, histidina, sacarose e água para injeção.

  - <b>Indicação:</b>

    - Para crianças e adolescentes, conforme recomendações das sociedades brasileiras de Pediatria (SBP) e Imunizações (SBIm).
    - Para adultos com até 50 anos, dependendo de risco epidemiológico.
    - Para viajantes com destino às regiões onde há risco aumentado da doença.
    - Para pessoas de qualquer idade com doenças que aumentem o risco para a doença meningocócica.


 - <b>Hepatite B</b> (<b>fonte:</b> *https://familia.sbim.org.br/vacinas/vacinas-disponiveis/vacina-hepatite-b*)

  - Previne contra infecção do fígado (hepatite) causada pelo vírus da hepatite B.

  - Trata-se de vacina inativada, portanto, não tem como causar a doença.

  - É composta por proteína de superfície do vírus da hepatite B purificado, hidróxido de alumínio, cloreto de sódio e água para injeção. Pode conter fosfato de sódio, fosfato de potássio e borato de sódio.

  - A rede pública utiliza a apresentação multidose (mais de uma dose por frasco), que contém timerosal (derivado do mercúrio) como conservante.

  - <b>Indicação:</b>

    - Para pessoas de todas as faixas etárias. Faz parte da rotina de vacinação das crianças, devendo ser aplicada, de preferência, nas primeiras 12-24 horas após o nascimento, para prevenir hepatite crônica – forma que acomete 90% dos bebês contaminados ao nascer.

    - Especialmente indicada para gestantes não vacinadas.


 - <b>Pneumocócica</b> (<b>fonte:</b> *https://familia.sbim.org.br/vacinas/vacinas-disponiveis/vacina-triplice-viral-sarampo-caxumba-e-rubeola-scr*)

  - Previne contra doenças causadas por 23 tipos de pneumococos.

  - Trata-se de vacina inativada, portanto, não tem como causar a doença.

  - É composta de partículas purificadas (polissacarídeos) das cápsulas de 23 tipos de Streptococos pneumoniae (pneumococo), cloreto de sódio, água para injeção e fenol.

  - <b>Indicação:</b>

    - Para crianças acima de 2 anos, adolescentes e adultos que tenham algum problema de saúde que aumenta o risco para doença pneumocócica (diabetes, doenças cardíacas e respiratórias graves; sem baço ou com o funcionamento comprometido desse órgão; com problemas de imunidade, entre outras condições).
    - Para pessoas a partir de 60 anos deve ser aplicada de rotina.
    - Não é recomendada como rotina para crianças, adolescentes e adultos saudáveis.

  - <b>Novas vacinas penumocócia</b>
    - Novas fórmulas de vacinas penumocócias foram implantadas no SUS. Matéria que explica como estão os novos imunizantes e suas recomendações. (<b>fontes:</b> *https://familia.sbim.org.br/vacinas/vacinas-disponiveis/vacinas-pneumococicas-conjugadas* , *https://saude.abril.com.br/medicina/mais-uma-vacina-contra-bacteria-causadora-da-pneumonia-chega-ao-sus/#:~:text=O%20Minist%C3%A9rio%20da%20Sa%C3%BAde%20anunciou,v%C3%ADrus%20HIV%20e%20indiv%C3%ADduos%20transplantados.*


 - <b>Poliomielite</b> (<b>fonte:</b> *https://familia.sbim.org.br/vacinas/vacinas-disponiveis/vacinas-poliomielite*)

  - Previne contra poliomielite (paralisia infantil).

  - Vacina Oral Poliomielite (VOP) – É uma vacina oral atenuada bivalente, ou seja, composta pelos vírus da pólio tipos 1 e 3, vivos, mas “enfraquecidos”. Contém ainda cloreto de magnésio, estreptomicina, eritromicina, polissorbato 80, L-arginina e água destilada.

  - Vacina Inativada Poliomielite (VIP) – Por ser inativada, não tem como causar a doença. É uma vacina trivalente e injetável, composta por partículas dos vírus da pólio tipos 1, 2 e 3. Contém ainda 2-fenoxietanol, polissorbato 80, formaldeído, meio Hanks 199, ácido clorídrico ou hidróxido de sódio. Pode conter traços de neomicina, estreptomicina e polimixina B, utilizados durante a produção.

  - <b>Indicação:</b>

    - Devido à erradicação da poliomielite em diversas regiões do mundo e também para evitar a paralisia que pode ser causada pelo vírus contido na vacina oral (VOP), a Organização Mundial da Saúde (OMS) recomenda que países como o Brasil, de baixo risco para o desenvolvimento da doença, passem a utilizar a vacina inativada (VIP), sempre que possível.
    - Desde 2016, o Programa Nacional de Imunizações (PNI) adota a vacina VIP nas três primeiras doses do primeiro ano de vida (aos 2, 4 e 6 meses de idade) e a VOP no reforço e campanhas anuais de vacinação.
    - A Sociedade Brasileira de Imunizações (SBIm) orienta que a VIP seja a vacina de preferência na administração de todas as doses.
    - A vacina poliomielite é indicada de rotina para todas as crianças menores de 5 anos.
    - Para viajantes adolescentes e adultos com destino a países onde a doença é endêmica, como o Paquistão e o Afeganistão, ou a locais onde há risco de transmissão e registro de casos de poliomielite causada pelo vírus vacinal.

Já a da Sanofi Pasteur contém: lactose, sorbitol, cloridrato de L-histidina, L-alanina e solução salina.

  - Trata-se de vacina atenuada, contendo vírus vivos “enfraquecidos” do sarampo, da rubéola e da caxumba; aminoácidos; albumina humana; sulfato de neomicina; sorbitol e gelatina. Contém também traços de proteína do ovo de galinha usado no processo de fabricação da vacina.

# Funções

**Função para gerar o df do(s) estado(s) escolhido(s) e gerar o barplot da(s) vacina(s) escolhidas no período de 2010 a 2019**

```python
def vacina_estado(estado: str):
  df_estado = df.loc[estado] # criando df com total de vacinas
  df_estado.reset_index(inplace=True) # resetando index
  df_estado.drop(['uf','Total'], axis=1, inplace=True) # dropando colunas uf e Total
  df_estado.set_index('ano', inplace=True) # setando ano como index
  return df_estado, estado
```

```python
def barplot_cv_estado(DataFrame, vacinas: list):

  estado = DataFrame[1] # recebendo o nome do estado do DataFrame
  if estado == 'Total': # se for Total mudar para Brasil
    estado = 'Brasil'

  DataFrame = DataFrame[0] # recebendo o DataFrame

  colunas_vac = vacinas # lista de vacinas recebida

  plt.figure(figsize=(18,15)) # tamanho da figura

  for i in enumerate(colunas_vac): # for loop para cada vacina passada
    plt.subplot(3,3, i[0]+1)

    # configurando a paleta de cores para barra de acordo com seu valor
    pal = sns.color_palette("flare_r", len(DataFrame[i[1]]))
    rank = DataFrame[i[1]].argsort().argsort()

    # criação do plot
    ax = sns.barplot(data=DataFrame, 
                x=DataFrame.index,
                y=i[1],
                palette=np.array(pal[::-1])[rank])
        
    # configurando a rotação do eixo x para 45 graus
    plt.xticks(rotation=45)

    # configurando o limite do eixo y
    plt.ylim(0,140)
    
    # configurando o eixo x
    plt.xlabel('')

    # configurando o grid
    plt.grid(True, color = '0.95', linestyle = '--')

    # Título superior
    plt.suptitle('Cobertura Vacinal - {} de 2010 a 2019'.format(estado).upper(), 
            weight='bold', fontsize=16)
    
    # salvando plot
    plt.savefig('barplot_{}'.format(estado), bbox_inches='tight', pad_inches=0.2)
```

**Função para gerar lineplot do(s) estado(s) e da(s) vacina(s) que deseja comparar**

```python
def plot_linha(estados: list, vacinas_pesquisa: list):
  
  df_linha = pd.DataFrame() # criando um obejto DataFrame

  for x in estados: # for loop para cada estado passado
    df_estado = vacina_estado(x)[0] # utilizando a função 'vacina_estado' que cria o DataFrame para o estado escolhido 
    df_estado['uf'] = vacina_estado(x)[1] # criando a coluna uf no DataFrame
    df_linha = df_linha.append(df_estado) # adicionando o DataFrame do estado escolhido criado no DataFrame principal
  
  vacinas_pesquisa = vacinas_pesquisa # recebendo a lista de vacinas
  vacinas_pesquisa.append('uf') # adicionando o item 'uf' na lista de vacinas para realizar o slicing corretamente depois
  
  df_linha = df_linha[vacinas_pesquisa] # selecionando no DataFrame as vacinas escolhidas
  
  df_linha.reset_index(inplace=True) # resetando o index

  vacinas = list(df_linha.columns.to_list())[1:-1] # selecionando as vacinas no DataFrame para realizar o plot

  plt.figure(figsize=(18,15)) # configurando o tamanho do plot

  for i in enumerate(vacinas): # for loop para as vacinas escolhidas
    sns.set_style("whitegrid") # configurando o estilo do plot

    plt.subplot(3,3,i[0]+1) # numero de subplots

    # criando o plot
    ax = sns.lineplot(data=df_linha, x='ano', y=i[1], 
                      hue='uf', 
                      palette=sns.color_palette("flare_r", len(set(list(df_linha['uf'])))))

    # rotacão dos ticks no eixo x
    plt.xticks(rotation=45)

    # configurando o grid
    plt.grid(True, color = '0.95', linestyle = '--')

    # configurando o label do eixo x
    plt.xlabel('')

    # configurando a quantidade de ticks no eixo x
    plt.locator_params(axis="x", nbins=10)


  # ajeitando a string para exibir título e salvar o arquivo
  s = ''
  for x in estados:
    s = s + '-' + x
  
  # Super título
  plt.suptitle(f'Lineplot Comparativo {s}'.upper(), weight='bold', fontsize=16)


  # salvando o plot
  plt.savefig('plot_linha_compara{}'.format(s), bbox_inches='tight', pad_inches=0.2)
```

**Função para gerar o df e plotar o heatmap**

```python
def df_heatmap(vacina: str):
  heatmap = df.reset_index()
  heatmap = heatmap.pivot('uf','ano', vacina)
  return heatmap, vacina
```

```python
def heatmap_plot(DataFrame):

  vacina = DataFrame[1]

  DataFrame = DataFrame[0]

  plt.figure(figsize=(9,7))

  ax = sns.heatmap(DataFrame, cmap="rocket_r", robust=True)
      
  plt.xticks(rotation=45)

  ax.set_xlabel('')

  ax.set_ylabel('')

  plt.title('{}\n'.format(vacina).upper(), weight='bold', fontsize=14)

  plt.savefig('{}'.format(vacina), bbox_inches='tight', pad_inches=0.2)
```

```python
def vacina_escolha_heatmap(vacinas: list):
  
  vac = vacinas

  for i in vac:
    heatmap_vac = df_heatmap(i)
    heatmap_plot(heatmap_vac)
```
