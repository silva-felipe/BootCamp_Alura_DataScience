## Desafio 04 -  Desenvolver uma função para a limpeza e tratamento dos dados, retornando duas tabelas "população" e "unida"

```python
import pandas as pd
pd.options.display.float_format = "{:.2f}".format

def populacao_estados():
  url = 'https://pt.wikipedia.org/wiki/Lista_de_unidades_federativas_do_Brasil_por_popula%C3%A7%C3%A3o#cite_note-IBGE_POP-1'
  url = 'http:' + url[6:]
  pop_estados = pd.read_html(url)[0]
  pop_estados = pop_estados.drop(['Posição','País comparável (habitantes)','% da pop. total'], axis=1)
  pop_estados = pop_estados.rename(columns={pop_estados.columns[0]:'uf', pop_estados.columns[1]:'populacao'})
  pop_estados = pop_estados.set_index(pop_estados.columns[0])
  pop_estados['populacao'] = pop_estados['populacao'].str.replace(' ','')
  pop_estados['populacao'] = pd.to_numeric(pop_estados['populacao'])
  

  url_processamento = 'https://raw.githubusercontent.com/silva-felipe/BootCamp_Alura_DataScience/main/dados/ano_mes_processamento.csv'
  df_processamento = pd.read_csv(url_processamento, encoding='ISO-8859-1',sep=';',skiprows=3,skipfooter=12,thousands='.',decimal=',')
  df_processamento['uf'] = df_processamento['Unidade da Federação'].str[3:]
  df_processamento.drop(['Unidade da Federação','Total'], axis=1, inplace=True)
  df_processamento.set_index('uf', inplace=True)
  gasto_do_mais_recente = df_processamento[df_processamento.columns[-1]]

  pop_gastos = pop_estados.join(gasto_do_mais_recente)
  
  return display(pop_estados), display(pop_gastos)


populacao_estados()
```

```python
meses = { 
    'Jan' : 1,
    'Fev' : 2,
    'Mar' : 3,
    'Abr' : 4,
    'Mai' : 5,
    'Jun' : 6,
    'Jul' : 7,
    'Ago' : 8,
    'Set' : 9,
    'Out' : 10,
    'Nov' : 11,
    'Dez' : 12
}

def para_dia(ano_mes: str):
    ano: int = int(ano_mes[:4])
    mes: str = ano_mes[5:]
    mes_numero: int = meses[mes]
    return date(ano, mes_numero, 1)
```
