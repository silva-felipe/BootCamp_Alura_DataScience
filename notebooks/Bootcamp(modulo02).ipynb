## Desafio video 01
ler diretamente do wikipedia (html)


```python
url='http://pt.wikipedia.org/wiki/Lista_de_unidades_federativas_do_Brasil_por_popula%C3%A7%C3%A3o#cite_note-IBGE_POP-1'
```

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_html.html
utilizando a documentação do pandas, temos a função pd.read_html. Nela eles dizem que se o link possuir https, devemos tirar o 's' para poder ler corretamente os dados.


```python
pd.read_html(url)[0]
```

limpando os dados

```python
pop_estados = pd.read_html(url)[0]
```

```python
pop_estados.head()
```

retirando as colunas indesejadas

```python
pop_estados = pop_estados.drop(['Posição','País comparável (habitantes)','% da pop. total'], axis=1)
```

```python
pop_estados.head()
```

verificando quais sao as colunas existentes e renomeando-as

```python
pop_estados.columns
```

```python
pop_estados = pop_estados.rename(columns={pop_estados.columns[0]:'uf', pop_estados.columns[1]:'populacao'})
```

```python
pop_estados.head()
```

setando um novo index

```python
pop_estados = pop_estados.set_index(pop_estados.columns[0])
```

```python
pop_estados.head()
```

retirando os espaços dos valores na coluna 'populacao' utilizando a função replace de python

```python
pop_estados['populacao'] = pop_estados['populacao'].str.replace(' ','')
```

transformando os valores de 'object' para 'int64' utilizando a função 'to_numeric'
fonte: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_numeric.html#pandas.to_numeric

```python
pop_estados['populacao'] = pd.to_numeric(pop_estados['populacao'])
```
